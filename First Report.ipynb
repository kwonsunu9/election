```python
import pandas as pd
import json, os 

with open('C:/apps/analysis/election/state_code.json', 'r', encoding='utf-8-sig') as json_file:
    json_data = json.load(json_file)

def country_name_convert(row):
    if row['state'] in json_data:
        return json_data[row['state']]
    return row['state']

def create_dateframe(filename):

    doc = pd.read_csv(PATH + filename, encoding='utf-8-sig')  # 1. csv 파일 읽기
    doc = doc[['state', 'candidatevotes']]  # 2. 특정 컬럼만 선택해서 데이터프레임 만들기
    doc = doc.dropna(subset=['candidatevotes'])     # 3. 특정 컬럼에 없는 데이터 삭제하기
    doc['state'] = doc.apply(country_name_convert, axis=1)   # 4. 'Country_Region'의 국가명을 여러 파일에 일관되게 변경하기
    doc = doc.astype({'candidatevotes': 'int64'})   # 5. 특정 컬럼의 데이터 타입 변경하기
    doc = doc.groupby('state').sum()  # 6. 특정 컬럼으로 중복된 데이터를 합치기

    # 7. 파일명을 기반으로 날짜 문자열 변환하고, 'Confirmed' 컬럼명 변경하기
    date_column = filename.split(".")[0]
    doc.columns = [date_column]
    return doc
```

```python
import os

def generate_dateframe_by_path(PATH):

    file_list, csv_list = os.listdir(PATH), list()
    first_doc = True
    for file in file_list:
        if file.split(".")[-1] == 'csv':
            csv_list.append(file)
    
    for file in csv_list:
        doc = create_dateframe(file)
        if first_doc:
            final_doc, first_doc = doc, False
        else:
            final_doc = pd.merge(final_doc, doc, how='outer', left_index=True, right_index=True)

    final_doc = final_doc.fillna(0)
    return final_doc
```

```python
def create_flag_link(row):
    flag_link = 'https://flagcdn.com/48x36/us-' + row + '.png'
    return flag_link
    
PATH = 'C:/apps/analysis/election/'
df_candidatevotes = generate_dateframe_by_path(PATH)
df_candidatevotes = df_candidatevotes.astype('int64')
state_info = pd.read_csv(PATH + "1976.csv", encoding='utf-8-sig')
state_info = state_info[['state', 'state_po']]
state_info = state_info.drop_duplicates(['state'], keep = 'first').groupby('state').sum()
state_info = state_info.rename(columns = {'state_po':'iso'})[['iso']]

doc_final_state = pd.merge(df_candidatevotes, state_info, how='left', on='state')
doc_final_state = doc_final_state.dropna(subset=['iso2'])
doc_final_state['iso2'] = doc_final_state['iso2'].apply(create_flag_link)

cols = doc_final_country.columns.tolist()
cols.remove('iso2')
cols.insert(1, 'iso2')
doc_final_country = doc_final_country[cols]
cols[1] = 'Country_Flag'
doc_final_country.columns = cols

doc_final_country.to_csv("election_graph.csv")
```
